```python
total_len = len(data['SeriousDlqin2yrs'])
percentage_labels = (data['SeriousDlqin2yrs'].value_counts()/total_len)*100
percentage_labels
```

```python
sns.set(style="whitegrid")
total_len = len(data)
plt.figure(figsize=(6, 4))
ax = sns.countplot(x=data['SeriousDlqin2yrs'], palette="viridis")
ax.set_title('Data Distribution of SeriousDlqin2yrs', fontsize=14)
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x() + p.get_width() / 2, 
            height + 2, 
            f'{100 * height / total_len:.2f}%', 
            fontsize=12, ha='center')

ax.set_xlabel("SeriousDlqin2yrs Labels", fontsize=12)
ax.set_ylabel("Number of Records", fontsize=12)

sns.set(font_scale=1.2)
plt.show()
```

```python
feature_x = "NumberOfDependents"
feature_y = "MonthlyIncome"

plt.figure(figsize=(8,5))
sns.scatterplot(x=data[feature_x], y=data[feature_y], color="blue", alpha=0.5)

plt.xlabel(feature_x)
plt.ylabel(feature_y)
plt.title(f"Scatter Plot of {feature_x} vs {feature_y}")

plt.show()
```

```python
corr_matrix = data.corr()

plt.figure(figsize=(8, 4))

sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)

plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
data = data.drop(['NumberOfTime60-89DaysPastDueNotWorse'],axis=1)
```

```python

feature =  'NumberOfDependents'  # Change this to your actual column name

# Create the boxplot
plt.figure(figsize=(8, 5))
sns.boxplot(x=data[feature])
plt.title(f"Boxplot of {feature} (Detecting Outliers)")
plt.show()
```

```python
q1= data[ 'NumberOfDependents'].quantile(0.25)
q3= data[ 'NumberOfDependents'].quantile(0.75)
iqr=q3-q1
```

```python
q1,q3,iqr
```

```python
upper_limit= q3+(1.5*iqr)
lower_limit= q1-(1.5*iqr)
lower_limit,upper_limit
```

```python
data.loc[(data['NumberOfDependents']>upper_limit)|(data['NumberOfDependents']<lower_limit)]
```

```python
new_data= data.loc[(data['NumberOfDependents']<upper_limit)&(data['NumberOfDependents']>lower_limit)]
print('before removing outliers:', len(data))
print('after removing outliers:', len(new_data))
print('outliers:', len(data)-len(new_data))
```

```python
plt.figure(figsize=(7, 4))
sns.boxplot(new_data['NumberOfDependents'])
plt.title(f"Boxplot of {feature} (Without Outliers)")
plt.show()
```
